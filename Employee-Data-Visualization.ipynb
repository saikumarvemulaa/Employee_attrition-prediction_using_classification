```python
import pandas as pd
```

```python
ibmdata = pd.read_csv("IBM data.csv")
```

```python
ibmdata
```

```python
#Filter the data of human resources
hribmdata = ibmdata[(ibmdata.Department == 'Human Resources')]
```

```python
hribmdata
```

```python
#Visualize the HR data
import matplotlib.pyplot as plt
```

```python
#Create and deploy a line chart

plt.figure(figsize=(20,10))
plt.plot(hribmdata['HourlyRate'])
plt.title("Line chart for hourly rate of employees of HR department")
```

```python
#Determine categories of job satisfaction
print("Categories of Job satisfaction:\n", hribmdata['JobSatisfaction'].unique())
```

```python
#Filtering records of different job satisfaction
satisfaction_1 = hribmdata['JobSatisfaction']==1
satisfaction_2 = hribmdata['JobSatisfaction']==2
satisfaction_3 = hribmdata['JobSatisfaction']==3
satisfaction_4 = hribmdata['JobSatisfaction']==4
```

```python
#Determining the number of records for different satisfaction levels
s1 = satisfaction_1.value_counts()
s2 = satisfaction_2.value_counts()
s3 = satisfaction_3.value_counts()
s4 = satisfaction_4.value_counts()
```

```python
#Creating a list of element
list =[s1[True],s2[True],s3[True],s4[True]]
print("The values are:\n", list)
```

```python
#Creating and displaying a pie chart
level = ['satisfaction_1','satisfaction_2','satisfaction_3','satisfacton_4']
plt.figure(figsize=(20,10))
plt.pie(list,labels=level)
plt.title("Job satisfaction of employees of HR department")
plt.show()
```

```python
#Create a scatterplot
plt.figure(figsize=(20,10))
plt.scatter(hribmdata['Age'],hribmdata['DailyRate'])
plt.title("Scatter plot for age with daily rate")
plt.show()
```

# Observation
There is no standard pattern between age and daily rates.
There are employees belonging to different age groups and having different daily rates.

```python
#Create multiple bar charts
#Filter data by Gender
Male = hribmdata[hribmdata['Gender'] == 'Male']
Female = hribmdata[hribmdata['Gender'] == 'Female']
```

```python
#Determine the mean of variables 
hribmdata.describe()
```

```python
#Determine the mean of variables for different gender
age_gender = [Male['Age'].mean(),Female['Age'].mean()]
daily_rate_gender = [Male['DailyRate'].mean(),Female['DailyRate'].mean()]
hourly_rate_gender = [Male['HourlyRate'].mean(),Female['HourlyRate'].mean()]
year_promotion_gender=[Male['YearsSinceLastPromotion'].mean(),Female['YearsSinceLastPromotion'].mean()]
year_company_gender=[Male['YearsAtCompany'].mean(),Female['YearsAtCompany'].mean()]
year_total_gender=[Male['TotalWorkingYears'].mean(),Female['TotalWorkingYears'].mean()]

```

```python
#Create a list of Gender
list = ['Male','Female']
```

```python
#Creating a chart
plt.figure(1,figsize=(20,10))
plt.suptitle("Information of employees of different genders")
plt.subplot(231)
plt.bar(list,age_gender)
plt.title("Age")
plt.subplot(232)
plt.bar(list,daily_rate_gender)
plt.title("Daily Rate")
plt.subplot(233)
plt.bar(list,hourly_rate_gender)
plt.title("Hourly Rate")
plt.subplot(234)
plt.bar(list,year_promotion_gender)
plt.title("Years since Last Promotion")
plt.subplot(235)
plt.bar(list,year_company_gender)
plt.title("Years at company")
plt.subplot(236)
plt.bar(list,year_total_gender)
plt.title("Total working years")
plt.show()
```

```python
#Create a box plot
#Filter the data on the basis of work-life balance
w1 = hribmdata[hribmdata['WorkLifeBalance']==1]
w2 = hribmdata[hribmdata['WorkLifeBalance']==2]
w3 = hribmdata[hribmdata['WorkLifeBalance']==3]
w4 = hribmdata[hribmdata['WorkLifeBalance']==4]
```

```python
#Creating a list of different sets of work-life balance
salary_hike=[w1['PercentSalaryHike'],w2['PercentSalaryHike'],w3['PercentSalaryHike'],w4['PercentSalaryHike']]
plt.figure(figsize=(10,4))
plt.title("Box plot for work life balance and percent salary hike")
plt.boxplot(salary_hike)
plt.show()
```

# Observation
1.There are four different categories for work-life balance:1,2,3 and 4.
2.The median of percent hike is least for employees having work-life balance as 4 or 2 and Maximum for employees having work-life balance as 1.
3.Beside the range of percent hike for employees having worklife balance as 1 is between 15 and 19.

```python
#Create  a stacked bar chart
plt.figure(1,figsize=(15,10))
Departments = ['Operations','Marketing','HR','Finance','Technical']
salary_1930=[10,20,24,30,20]
salary_1940=[12,27,30,42,26]
salary_1950=[15,35,34,50,33]
salary_1960=[27,43,43,57,38]
salary_1970=[35,45,46,62,40]
salary_1980=[38,49,55,65,45]
salary_1990=[42,52,60,72,50]
salary_2000=[48,54,65,75,58]
salary_2010=[53,58,70,77,63]

plt.bar(Departments,salary_2010,color='green')
plt.bar(Departments,salary_2000,color='red')
plt.bar(Departments,salary_1990,color='yellow')
plt.bar(Departments,salary_1980,color='blue')
plt.bar(Departments,salary_1970,color='brown')
plt.bar(Departments,salary_1960,color='orange')
plt.bar(Departments,salary_1950,color='purple')
plt.bar(Departments,salary_1940,color='magenta')
plt.bar(Departments,salary_1930,color='pink')

#Add labels and title to the chart
plt.xlabel('Departments')
plt.ylabel('Salary in Different Years')
plt.title('Stacked bar chart')
plt.show()
```
